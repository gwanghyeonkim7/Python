```python
class Car:
    def __init__(self):
        self.car = []
    def add_option(self, option):
        self.car.append(option)       
            
    def show_option(self):
        return self.car
        
        
car1 = Car()
car2 = Car()
car1.add_option('전동 트렁크')
car1.add_option('통풍 시트')
car2.add_option('뒷자리 에어백')
print(car1.show_option)
```

```python
class EachSum:
    def __init__(self):
        self.total = 0
    
    def compute(self, num):
        while True:
            self.total += num% 10
            num //= 10
            if num == 0: break
                
        return self.total


        
        # 코드 작성
each = EachSum()
total = each.compute(12345)
print('합계', total)
```

```python
import random
class SelectMenu:
    def __init__(self, menu):
        self.menu = menu
    def get_menu(self):
        num = random.randint(0, len(self.menu)-1)
        return self.menu(num)
    
menu = SelectMenu(['짬뽕', '초밥', '쌀국수', '주꾸미'])
print(menu.get_menu())
```

```python
file = open('C:/dev/[ex]1-document.txt', 'rt')
text = file.read()
text = text.replace("\n", " ")
words = text.split(" ")
# Shift + Ctrl + -
```

```python
word_dict = dict()
for word in words:
    if word in word_dict:
        word_dict[word] = word_dict[word] + 1
    else:
        word_dict[word] = 1
word_dict
```

```python
import operator

word_list = sorted(
    word_dict.items(), 
    key=operator.itemgetter(1), reverse=True)
```

```python
for key, value in word_list[:5]:
    print(key, value)
```
